```python
from dataclasses import dataclass
import sys
sys.path.append('../estimator')

from estimator import LWE, ND
from estimator.nd import sigmaf,stddevf
from sage.all import var, log, ceil, floor, sqrt, Infinity, euler_phi, round, pi, n, ln
import os

class HiddenPrints:
    def __enter__(self):
        self._original_stdout = sys.stdout
        sys.stdout = open(os.devnull, "w")

    def __exit__(self, exc_type, exc_val, exc_tb):
        sys.stdout.close()
        sys.stdout = self._original_stdout

@dataclass
class RBEParams:
    """
    Dataclass for the parameters of the RBE.
    """

    secpar: int # security level
    f: int # conductor
    n: int # module rank, height of matrix A 
    l_A: int # m_A = n * l_A
    m_A: int # width of matrix A
    B: int # number of dictionaries
    q: int # modulus
    sigma_tilde: float # Gaussian parameter
    D: int # number of bits dropped

    def __repr__(self):
        return f'RBE( secpar: {self.secpar:3d}, f: {self.f:3d}, n: {self.n:1d}, m_A: {self.m_A:1d}, q: 2^{ceil(log(self.q,2)):3d}, sigma_tilde: {self.sigma_tilde:5.1f}, (|ct|: {self.ct():5s} kB)'

    def ct(self):
        # return ciphertext size in KB
        a = 162 * 3 * self.m_A
        b = self.n
        c = self.B
        return f'{n( (a + b + c) * self.f/2 * (ceil(log(self.q, 2)) - self.D) / (1<<13)):5.1f}'

def rbe_params(secpar, f, n, l_A, sigma):
    # static
    l_hat = 162
    B = 30
    
    # Table 2
    phi = euler_phi(f)
    m_A = n * l_A
    k = 3

    # heuristic correctness, remark 4.4
    b = ceil((7 * l_hat * sqrt(phi**3 * l_A * n) * sigma)**(1/l_A))

    D = round(log( b**(l_A + 1) / (4 * (1 + (l_hat +1) * b)), 2))
    added_bit_dropping_term = 4 * 2**D * (1 + (l_hat + 1) * b)
    q = ceil(b**(l_A + 1)) + added_bit_dropping_term
    
    # q = 4 * ( (3 * l_A + 1) * l_hat + 2) * phi**2 * n * sigma * b + 2
    sigma_tilde = 1.7 * l_hat * phi * sqrt(l_A * n) * sigma * b

    # security, corollary 4.4
    smoothing_param = sqrt( ln( 2 * phi * (k * n * l_A + B) * (1 + 2**secpar) ) / pi )
    # eq = sqrt( (4 * sigma_tilde**2) / ( (k * l_A + 1) * n * B * phi**2 * b**2 + sigma**(-2) ) - 2 * smoothing_param**2 )
    sigma_star = sqrt(((k * l_A + 1) * n * B * phi**2 * b**2 / (4 * sigma_tilde**2) + sigma**(-2))**(-1) - 2 * smoothing_param**2)

    assert sigma_star >= sqrt(2) * smoothing_param
        
    lwe1 = LWE.Parameters(
        n=phi*n, m = phi*n, q=q, Xs=ND.Uniform(-b/2, b/2), Xe=ND.Uniform(-b/2, b/2)
    )
    lwe2 = LWE.Parameters(
        n=phi*n, m = phi * (k * n * l_A + B), q=q, Xs=ND.DiscreteGaussian(stddevf(sigma_star)), Xe=ND.DiscreteGaussian(stddevf(sigma_star))
    )
    
    lwe_params = [
        lwe1, 
        lwe2,
    ]
    
    with HiddenPrints():
        lwe_costs = []
        for lwe in lwe_params:
            costs = LWE.estimate(lwe, deny_list=("arora-gb", "bkw", "bdd_hyrbid", "bdd_mitm_hybrid"))
            print(costs)
            lwe_costs += [costs]
    
    min_lwe_costs = []
    for costs in lwe_costs:
        min_cost = log(min(cost["rop"] for cost in costs.values()),2)
        min_cost = floor(min_cost) if min_cost < Infinity else Infinity
        min_lwe_costs += [min_cost]

    min_lwe_cost = min(min_lwe_costs)

    print(f'secpar: {min_lwe_costs}, n: {n}, b: {b}, q: 2^{ceil(log(q,2)):3d}, D: {D}')
               
    if (
        min_lwe_cost >= secpar 
    ):
        rbe = RBEParams(min_lwe_cost, f, n, l_A, m_A, B, q, sigma_tilde.n(), D)
        print(f"{rbe}")
```

```python
rbe_params(secpar=128, f=2048, n=2, l_A=2, sigma=2**4)
```

```python
# Concrete parameter set
c_secpar = 128
c_phi = 512
c_n = 3
c_k = 3
c_l_A = 2
c_B = 30
c_q = 217579865116673
c_l_hat = 162
c_sigma = 64
c_b = 77
c_sigma_tilde = 100_000_000_000

c_smoothing_param = sqrt( ln( 2 * c_phi * (c_k * c_n * c_l_A + c_B) * (1 + 2**c_secpar) ) / pi )
c_sigma_star = sqrt(((c_k * c_l_A + 1) * c_n * c_B * c_phi**2 * c_b**2 / (4 * c_sigma_tilde**2) + c_sigma**(-2))**(-1) - 2 * c_smoothing_param**2)

c_lwe1 = LWE.Parameters(
    n=c_phi*c_n, m = c_phi*c_n, q=c_q, Xs=ND.Uniform(-c_b/2, c_b/2), Xe=ND.Uniform(-c_b/2, c_b/2)
)
c_lwe2 = LWE.Parameters(
    n=c_phi*c_n, m = c_phi * (c_k * c_n * c_l_A + c_B), q=c_q, Xs=ND.DiscreteGaussian(stddevf(c_sigma_star)), Xe=ND.DiscreteGaussian(stddevf(c_sigma_star))
)

LWE.estimate(c_lwe1, deny_list=("arora-gb", "bkw", "bdd_hyrbid", "bdd_mitm_hybrid"))
_ = LWE.estimate(c_lwe2, deny_list=("arora-gb", "bkw", "bdd_hyrbid", "bdd_mitm_hybrid"))
```
