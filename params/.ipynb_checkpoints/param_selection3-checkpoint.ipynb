```python
from dataclasses import dataclass
import sys
sys.path.append('../estimator')

from estimator import LWE, ND
from estimator.nd import sigmaf,stddevf
from sage.all import var, log, ceil, floor, sqrt, Infinity, euler_phi, round, pi, n, ln
import os

class HiddenPrints:
    def __enter__(self):
        self._original_stdout = sys.stdout
        sys.stdout = open(os.devnull, "w")

    def __exit__(self, exc_type, exc_val, exc_tb):
        sys.stdout.close()
        sys.stdout = self._original_stdout

@dataclass
class RBEParams:
    """
    Dataclass for the parameters of the RBE.
    """

    secpar: int # security level
    f: int # conductor
    n: int # module rank, height of matrix A 
    l_A: int # m_A = n * l_A
    m_A: int # width of matrix A
    B: int # number of dictionaries
    q: int # modulus
    sigma_tilde: float # Gaussian parameter
    D: int # number of bits dropped

    def __repr__(self):
        return f'RBE( secpar: {self.secpar:3d}, f: {self.f:3d}, n: {self.n:1d}, m_A: {self.m_A:1d}, q: 2^{ceil(log(self.q,2)):3d}, sigma_tilde: {self.sigma_tilde:5.1f}, (|ct|: {self.ct():5s} MB)'

    def ct(self):
        # return ciphertext size in KB
        a = 162 * 3 * self.m_A
        b = self.n
        c = self.B
        return f'{n( (a + b + c) * self.f/2 * (ceil(log(self.q, 2)) - self.D) / (1<<23)):5.3f}'

def rbe_params(secpar, f, n, l_A, sigma):
    # static
    l_hat = 162
    B = 30
    
    # Table 2
    phi = euler_phi(f)
    m_A = n * l_A
    k = 3

    # remark 4.4 before application of heuristics
    b = ceil((12 * l_hat * phi**(3/2) * l_A * n * sigma)**(1 / l_A))
    sigma_tilde = 3 * l_hat * phi * l_A * n * sigma * b
    # correctness, cor. 4.3
    q_without_bit_dropping = 4 * ((3 * l_A + 1) * l_hat + 2) * phi**(3/2) * n * sigma * b + 4 * sqrt(phi) * sigma_tilde + 2
    D = round(log( q_without_bit_dropping / (4 * (1 + (l_hat +1) * b)), 2))
    added_bit_dropping_term = 4 * 2**D * (1 + (l_hat + 1) * b)
    q = q_without_bit_dropping + added_bit_dropping_term

    # security, corollary 4.4
    smoothing_param = sqrt( ln( 2 * phi * (k * n * l_A + B) * (1 + 2**secpar) ) / pi )
    # eq = sqrt( (4 * sigma_tilde**2) / ( (k * l_A + 1) * n * B * phi**2 * b**2 + sigma**(-2) ) - 2 * smoothing_param**2 )
    sigma_star = sqrt(((k * l_A + 1) * n * B * phi**2 * b**2 / (4 * sigma_tilde**2) + sigma**(-2))**(-1) - 2 * smoothing_param**2)

    assert sigma_star >= sqrt(2) * smoothing_param
        
    lwe1 = LWE.Parameters(
        n=phi*n, m = phi*n, q=q, Xs=ND.Uniform(-b/2, b/2), Xe=ND.Uniform(-b/2, b/2)
    )
    lwe2 = LWE.Parameters(
        n=phi*n, m = phi * (k * n * l_A + B), q=q, Xs=ND.DiscreteGaussian(stddevf(sigma_star)), Xe=ND.DiscreteGaussian(stddevf(sigma_star))
    )
    
    lwe_params = [
        lwe1, 
        lwe2,
    ]
    
    with HiddenPrints():
        lwe_costs = []
        for lwe in lwe_params:
            costs = LWE.estimate(lwe, deny_list=("arora-gb", "bkw", "bdd_hyrbid", "bdd_mitm_hybrid"))
            print(costs)
            lwe_costs += [costs]
    
    min_lwe_costs = []
    for costs in lwe_costs:
        min_cost = log(min(cost["rop"] for cost in costs.values()),2)
        min_cost = floor(min_cost) if min_cost < Infinity else Infinity
        min_lwe_costs += [min_cost]

    min_lwe_cost = min(min_lwe_costs)

    print(f'secpar: {min_lwe_costs}, n: {n}, b: {b}, q: {int(round(q))} ≈ 2^{log(q,2).n():5.1f}, D: {D}')
               
    if (
        min_lwe_cost >= secpar 
    ):
        rbe = RBEParams(min_lwe_cost, f, n, l_A, m_A, B, q, sigma_tilde, D)
        print(f"{rbe}")
```

```python
rbe_params(secpar=128, f=2048, n=2, l_A=2, sigma=2**4)
```

```python
rbe_params(secpar=128, f=1024, n=4, l_A=2, sigma=2**4)
```
